```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import HTML
from sklearn.datasets import load_iris
```

# Load data

```python
iris = load_iris()
```

```python
# Print dataset description
# print(iris.DESCR)
```

```python
X = iris.data
X[:10]
```

```python
y = iris.target
y
```

```python
classes = iris.target_names
classes
```

```python
feature_names = iris.feature_names
feature_names
```

# Visualize data

```python
html1 = '<center><img src="iris_classes.png"/></center>'
HTML(html1)
```

```python
html2 = '<center><img src="iris_petal_sepal.jpg"/></center>'
HTML(html2)
```

```python
ax = sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, palette='deep').set(
    title='Sepal', xlabel='Length [cm]', ylabel='Width [cm]')
plt.legend(labels=classes, loc='lower right')
plt.show()
```

```python
sns.scatterplot(x=X[:,2], y=X[:,3], hue=y, palette='deep').set(
    title='Petal', xlabel='Length [cm]', ylabel='Width [cm]')
plt.legend(labels=classes, loc='lower right')
plt.show()
```

It seems that:
1. Iris setosa will be the easiest to distinguish
2. Petal size is a more significant feature
